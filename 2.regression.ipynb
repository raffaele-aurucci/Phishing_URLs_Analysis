## **Correlation with label**

```r
if (!require(ggplot2))
    install.packages("ggplot2")

if (!require(corrplot))
    install.packages("corrplot")

library(ggplot2)
library(corrplot)
```

### Pearson's Correlation Matrix

```r
corr_pearson <- cor(df, method = "pearson")
```

```r
# corrplot dimension
options(repr.plot.width = 15, repr.plot.height = 10)

corrplot(corr_pearson,
         method = "circle",
         type = "upper",
         col = colorRampPalette(c("blue", "white", "red"))(200),
         cex.main = 1.75,
         tl.col = "black",
         tl.srt = 45,
         addCoef.col = "black",
         number.cex = 0.7,
         cl.pos = "r",
         tl.cex = 1.2,
)
```

### Spearman's Correlation Matrix

```r
corr_spearman <- cor(df, method = "spearman")
```

```r
# corrplot dimension
options(repr.plot.width = 15, repr.plot.height = 10)

corrplot(corr_spearman,
         method = "circle",
         type = "upper",
         col = colorRampPalette(c("blue", "white", "red"))(200),
         mar = c(0, 0, 1.75, 0),
         cex.main = 1.75,
         tl.col = "black",
         tl.srt = 45,
         addCoef.col = "black",
         number.cex = 0.7,
         cl.pos = "r",
         tl.cex = 1.2,
)
```
